```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,explained_variance_score
from sklearn.ensemble import ExtraTreesRegressor
```

```python
diamondsp=pd.read_csv("predict.csv")
diamondst=pd.read_csv("train.csv")

diamondst.id.count()
```

```python
le = preprocessing.LabelEncoder()
diamondsp["numclarity"] = le.fit_transform(diamondsp["clarity"])
diamondsp["numcolor"]=le.fit_transform(diamondsp["color"])
diamondsp["numcut"]=le.fit_transform(diamondsp["cut"])
diamondsp.head()
```

```python

le = preprocessing.LabelEncoder()
diamondst["numclarity"] = le.fit_transform(diamondst["clarity"])
diamondst["numcolor"]=le.fit_transform(diamondst["color"])
diamondst["numcut"]=le.fit_transform(diamondst["cut"])
diamondst.head()
```

```python
y_train=diamondst["price"]
X_train=diamondst[["carat","numclarity","numcolor","numcut","depth","table","x","y","z"]]
X_test=diamondsp[["carat","numclarity","numcolor","numcut","depth","table","x","y","z"]]
```

```python
from sklearn.model_selection import cross_validate

model2=ExtraTreesRegressor(n_estimators=350,max_features=9,max_depth=35,random_state=120,min_samples_split=9)
model2=model2.fit(X_train,y_train)
y_pred= model2.predict(X_test)

```

```python
diamondsp["price"]=y_pred
diamondsp
```
